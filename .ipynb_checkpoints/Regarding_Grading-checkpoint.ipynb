```python
# Remove unnecessary columns
columns_to_drop = ['Student_ID', 'First_Name', 'Last_Name', 'Email']
df = df.drop(columns=columns_to_drop, errors='ignore')  
```

#### 2. Check Null Values

```python
# Get missing values count for each column
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Fill missing numerical values with median
df["Attendance (%)"] = df["Attendance (%)"].fillna(df["Attendance (%)"].median())
df["Assignments_Avg"] = df["Assignments_Avg"].fillna(df["Assignments_Avg"].median())

# Fill missing categorical values
df["Parent_Education_Level"] = df["Parent_Education_Level"].fillna("Unknown")
```

```python
# Get missing values count for each column
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['Engagement'] = df['Attendance (%)'] * df['Participation_Score'] # Identifying the engagements of acedamic activities and engagement with letureres
df['is_stressed'] = (df['Stress_Level (1-10)'] > 7).astype(int) # finding are students having stressed
df['low_sleep'] = (df['Sleep_Hours_per_Night'] < 5).astype(int) # finding students having low sleep
```

```python
df['Focus_Quizzes'] = (df['Quizzes_Avg'] < 70).astype(int)
df['Focus_Participation'] = (df['Participation_Score'] < 6).astype(int)
df['Focus_Assignments'] = (df['Assignments_Avg'] < 70).astype(int)
df['Focus_Projects'] = (df['Projects_Score'] < 70).astype(int)
```

```python
# Checking for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

#### 5. Identifying the Distribution of the data, and select the target variable

```python
# Helper to get non-binary numeric columns
def get_non_binary_numerical_columns(data):
    numeric_cols = data.select_dtypes(include=['int64', 'float64'])
    return numeric_cols.loc[:, numeric_cols.nunique() > 2]

# 1. Distribution Plots
def plot_distripution(data):
    numerical_cols = get_non_binary_numerical_columns(data).columns
    histo_num_features = len(numerical_cols)
    histo_rows = (histo_num_features // 3) + (histo_num_features % 3 > 0)
    histo_cols = min(3, histo_num_features)

    fig, axes = plt.subplots(histo_rows, histo_cols, figsize=(12, 8))
    axes = axes.flatten()

    for i, col in enumerate(numerical_cols):
        sns.histplot(data[col], bins=20, kde=True, ax=axes[i])
        axes[i].set_title(f'Distribution of {col}')

    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# 2. Outlier Detection
def check_outliers_by_boxplot(data):
    numeric_columns = get_non_binary_numerical_columns(data).columns

    plt.figure(figsize=(12, 6))
    sns.boxplot(data=data[numeric_columns], showfliers=True)
    plt.xticks(rotation=45)
    plt.title("Boxplot for Non-Binary Numeric Columns")
    plt.show()

# 3. Correlation Matrix
def plot_correlation_matrix(data):
    numeric_data = get_non_binary_numerical_columns(data)
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title("Correlation Matrix of Non-Binary Numeric Columns")
    plt.show()

# 4. Scatter Matrix
def plot_scatter_matrix_features(data, target, important_features):
    scatter_num_features = len(important_features)
    scatter_rows = (scatter_num_features // 3) + (scatter_num_features % 3 > 0)
    scatter_cols = min(3, scatter_num_features)

    fig, axes = plt.subplots(scatter_rows, scatter_cols, figsize=(15, 20))
    axes = axes.flatten()

    for i, feature  in enumerate(important_features):
        sns.scatterplot(x=data[feature], y=data[target], ax=axes[i], alpha=0.5)
        axes[i].set_title(f'{feature} vs {target}')

    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# 5. Pie Chart (no need to change - it's for categorical)
def plot_pie_chart(data, column, title):
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(
        kind='pie', autopct='%1.1f%%', startangle=90, cmap='coolwarm', wedgeprops={'edgecolor': 'black'}
    )
    plt.title(title)
    plt.ylabel('')
    plt.show()

# 6. Clustered Distribution
def plot_clustered_distribution(data, feature, group_by_col, xlabel, title_count, title_percentage):
    plt.figure(figsize=(12, 6))
    categories = data[group_by_col].unique()
    colors = sns.color_palette("tab10", len(categories))

    feature_counts = data.groupby([feature, group_by_col]).size().unstack(fill_value=0)
    width = 0.15
    x = np.arange(len(feature_counts))

    fig, ax = plt.subplots(figsize=(12, 6))

    for i, category in enumerate(categories):
        ax.bar(x + i * width, feature_counts[category], width=width, label=category, color=colors[i], alpha=0.8, edgecolor="black")

    ax.set_xticks(x + width * (len(categories) / 2))
    ax.set_xticklabels(feature_counts.index, rotation=0)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Number of Students")
    ax.set_title(title_count)
    ax.legend(title=group_by_col)
    ax.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()

    total_students_per_category = data[group_by_col].value_counts()
    relative_feature = feature_counts.div(total_students_per_category) * 100

    fig, ax = plt.subplots(figsize=(12, 6))
    for i, category in enumerate(categories):
        ax.bar(x + i * width, relative_feature[category], width=width, label=category, color=colors[i], alpha=0.8, edgecolor="black")

    ax.set_xticks(x + width * (len(categories) / 2))
    ax.set_xticklabels(relative_feature.index, rotation=0)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Percentage of Students")
    ax.set_title(title_percentage)
    ax.legend(title=group_by_col)
    ax.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()
```

## Model Building

### 1. Classification Model 
This can be used to Identify the students, what are the things which want to develop according to their life with education patterns, 'Focus_Quizzes','Focus_Participation','Focus_Assignments','Focus_Projects'

```python
# Identify numeric columns, excluding specific focus flags
clf_flags = {'Focus_Quizzes','Focus_Participation','Focus_Assignments','Focus_Projects'}
numeric_cols = df.select_dtypes(include=[np.number]).columns
numeric_cols = [col for col in numeric_cols if col not in clf_flags]  

# Let's define these purely for classification
clf_features = [col for col in df.columns 
                if col not in clf_flags and col != 'Final_Score']  # exclude Final_Score (the regression target)

X_clf_all = df[clf_features]
clf_targets = ['Focus_Quizzes','Focus_Participation','Focus_Assignments','Focus_Projects']
```

##### In this section, we define a classification pipeline where we standardize all numeric columns using the StandardScaler. This is done to ensure that each feature contributes equally to the model by scaling them to have zero mean and unit variance. Since label encoding and one-hot encoding have already been applied to categorical columns, we focus on scaling the numeric features for model training. This simple approach ensures that the data is well-prepared for classification tasks.

```python
# Here, everything is numeric now, but let's separate "Final_Score" if it slipped in
num_cols_for_clf = [col for col in X_clf_all.columns if X_clf_all[col].dtype in [np.float64, np.int64]]
# We might do 'Final_Score' for regression only, so ensure it's excluded from classification features:
if 'Final_Score' in num_cols_for_clf:
    num_cols_for_clf.remove['Final_Score', 'Grade', 'Total_Score']

preprocessor_clf = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), num_cols_for_clf),
    ],
    remainder='passthrough'  
)
```

```python
pca_clf = PCA(n_components=0.95)  # dimension reduction
```

```python
# Stacking classifier with 2 base learners
base_classifiers = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('xgb', XGBClassifier(eval_metric='logloss', random_state=42))
]

final_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

stack_clf = StackingClassifier(
    estimators=base_classifiers,
    final_estimator=final_classifier,
    passthrough=True
)
```

##### Finding the Class imbalances and training

```python
# We'll wrap this in an imbalanced pipeline with SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.impute import SimpleImputer

for target in clf_targets:
    print(f"\n{'='*60}")
    print(f"Classification for {target}")
    print(f"{'='*60}")

    y_clf = df[target]

    # Train/test split (stratify on the target)
    X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
        X_clf_all, y_clf, test_size=0.2, random_state=42, stratify=y_clf
    )

    # Construct pipeline
    clf_pipeline = ImbPipeline([
        ('preprocessor', preprocessor_clf),
        ('smote', SMOTE(random_state=42)),
        ('imputer_post_smote', SimpleImputer(strategy='median')),
        ('pca', pca_clf),
        ('classifier', stack_clf)
    ])

    clf_pipeline.fit(X_train_clf, y_train_clf)
    y_pred_clf = clf_pipeline.predict(X_test_clf)

    # Evaluate
    print("Classification Report:")
    print(classification_report(y_test_clf, y_pred_clf))
    acc = accuracy_score(y_test_clf, y_pred_clf)
    f1 = f1_score(y_test_clf, y_pred_clf, average='binary')
    print(f"Accuracy: {acc:.3f}, F1: {f1:.3f}")

    # Confusion matrix
    cm = confusion_matrix(y_test_clf, y_pred_clf)
    plt.figure()
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {target}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    # ROC-AUC
    y_prob = clf_pipeline.predict_proba(X_test_clf)[:, 1]
    fpr, tpr, _ = roc_curve(y_test_clf, y_prob)
    auc_val = roc_auc_score(y_test_clf, y_prob)

    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC = {auc_val:.3f}")
    plt.plot([0,1],[0,1],'k--')
    plt.title(f"ROC Curve - {target}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Save model
    model_filename = "Final_Model_classifier.pkl"
    joblib.dump(clf_pipeline, model_filename)
    print(f"Saved classification model to {model_filename}")
```
